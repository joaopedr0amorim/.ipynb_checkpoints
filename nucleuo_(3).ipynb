```python
# Criar banco de dados SQLite
from sqlalchemy import create_engine, inspect
banco_de_dados = create_engine("sqlite:///dados/boletim")
banco_de_dados
```

```python
conexao = banco_de_dados.connect()
```

```python
# Importar tabela 6588 do Sidra/IBGE
dados_cana = sidra.get_table(
    table_code = "6588",
    territorial_level = "3",
    ibge_territorial_code = "27",
    variable = "35",
    period = "all",
    classifications = {"48": "39456"}
    )
```

```python
# Armazenar dados do exel como uma tabela a ser criada no banco SQL
dados_cana.to_sql(name = "cana", con = conexao, index = False)
```
